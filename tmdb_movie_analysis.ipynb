```python
original_language_counts = tmdb["original_language"].value_counts()
original_language_counts
```

```python
original_language_counts.index
```

```python
original_language_counts.values
```

```python
original_language_frame = original_language_counts.to_frame().reset_index()
original_language_frame.columns = ["original_language", "total"]
original_language_frame
original_language_frame.sort_values("total", ascending=False)
```

```python
sns.barplot(x="original_language", y="total", data=original_language_frame)
```

```python
sns.countplot(x="original_language", data=tmdb)
```

```python
# ☠️

original_language_frame.plot(kind="pie", y="total", labels=original_language_frame["original_language"])
```

```python
total_english_movies = original_language_counts["en"]
total_all_movies = original_language_counts.sum()
total_all_but_english_movies = total_all_movies - total_english_movies

language_data = {
    "language": ["English", "Other"],
    "total": [total_english_movies, total_all_but_english_movies]
}

language_frame = pd.DataFrame(language_data)
language_frame
```

```python
language_frame.plot(kind="pie", y="total", labels=language_frame["language"])
```

```python
sns.barplot(x="language", y="total", data=language_frame)
```

```python
sns.color_palette("mako", as_cmap=True)
```

```python
import matplotlib.pyplot as plt

all_movies_without_english = tmdb.query("original_language != 'en'")
all_movies_without_english_order = all_movies_without_english["original_language"].value_counts().index

plt.figure(figsize=(12, 6))
sns.countplot(
    x="original_language", # x-axis
    data=all_movies_without_english, # data
    order=all_movies_without_english_order, # order the x-axis it can be a list of strings or a Series
    hue="original_language", # color the points by the original language
    hue_order=all_movies_without_english_order, # order the hue it can be a list of strings or a Series
    palette="mako", # collor palette,
    stat="percent", # show the percentage of the points
    
)
plt.title("Movies without English")
plt.show()
```

```python
fr_es_de_with_revenue_and_budget = tmdb.query(
    "original_language in ['fr', 'es', 'de'] and revenue > 0 and budget > 0"
)
sns.lmplot(
    x="budget", # x-axis
    y="revenue", # y-axis
    data=fr_es_de_with_revenue_and_budget, # data
    col="original_language", # split the plot by the original language
    hue="original_language", # color the points by the original language
    ci=None, # confidence interval
    height=4, # height 
    scatter_kws={"s": 30, "alpha": 0.5}, # scatter plot size and transparency
)
```
